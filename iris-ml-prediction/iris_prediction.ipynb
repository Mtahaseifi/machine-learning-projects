```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.columns, df.info
```

```python
count_by_species = df['species'].value_counts().to_dict()
plt.bar(count_by_species.keys(),count_by_species.values())
plt.xlabel('species')
plt.ylabel('Numbers')
```

```python
label_encoder_manual = LabelEncoder()
df['species'] = label_encoder_manual.fit_transform(df['species'])
```

```python
x = df.drop(columns=['species'])
y = df['species']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=3 , random_state=42)
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
f'accuracy of model : {accuracy}'
```

```python
confusion_metrix = confusion_matrix(y_test , y_pred)
displaymatrix = ConfusionMatrixDisplay(confusion_metrix , display_labels=['setosa' , 'versicolor' , 'virginica'])
displaymatrix.plot(cmap='gray')
plt.title('confusion matrix')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plot_tree(model,feature_names=x.columns,class_names=['setosa' , 'versicolor' , 'virginica'], filled=True, fontsize=10,)
plt.title('algorithm of iris prediction model ')
plt.show()
```
